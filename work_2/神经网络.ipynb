```python
#导入sklearn包
import sklearn
```

```python
#导入所需要的包
from sklearn import datasets
```

```python
#导入数据
iris = datasets.load_iris()

print(iris)
```

```python
#训练集、测试集划分

from sklearn import model_selection
train, test, t_train, t_test = model_selection.train_test_split(iris['data'], iris['target'], test_size=0.3, random_state=999)
```

```python
#输出训练集、测试集，以及训练集、测试集标签的维度
print(train.shape)
print(test.shape)
print(t_train.shape)
print(t_test.shape)
```

```python
#神经网络
from sklearn.neural_network import MLPClassifier
ann = MLPClassifier(hidden_layer_sizes=2,max_iter=10000, activation="tanh")
ann = ann.fit(train,t_train)
```

```python
#对测试集分类

res_ann = ann.predict(test)
```

```python
print(res_ann)
```

```python
#对结果进行评估
from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(t_test,res_ann)
confmat
```

```python
#对结果进行评估
from sklearn.metrics import classification_report
print(classification_report(t_test,res_ann))
```
